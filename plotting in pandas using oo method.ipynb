```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

# Plotting with pandas using the OO method
- For simple and quick plots use Pyplot ie, something.plot()
- For more complicated plots, you'll want to use the OO method.

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
# Perform data analysis on patients over 50
over50 = heart_disease[heart_disease["age"]>50]
over50.head()
```

```python
#pyplot method
#c means colour and target contains 1's and 0's
over50.plot(kind="scatter",x="age",y="chol",c="target");
```

```python
#OO method mixed with Pyplot
#if we notice in this method the labels get predefined
fig,ax =plt.subplots(figsize=(10,6))
over50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax);
```

```python
#OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))
#plot the data
scatter = ax.scatter(x= over50["age"],
                        y = over50["chol"],
                        c = over50["target"])
#customize the plot
ax.set(xlabel = "Age", ylabel = "Cholestrol" , title = "Heart Disease and Cholestrol Levels")
#add legend
# unpacks the data from scatter and look at the target ans we are plotting target values
ax.legend(*scatter.legend_elements(),title = "Target");
#add a horizontal line
ax.axhline(over50["chol"].mean(),linestyle="--");
```

```python
fig , (ax1,ax2) = plt.subplots(figsize = (12,11),sharex=True, nrows=2,ncols=1)

scatter1 = ax1.scatter(x= over50["age"], y = over50["chol"], c = over50["target"])
scatter2 = ax2.scatter(x= over50["age"], y = over50["thalach"], c = over50["target"])

ax1.set(ylabel = "Cholestrol" , title = "Heart Disease and Cholestrol")
ax2.set(ylabel = "Max heart rate achieved" , title = "Heart Disease and Max heart rate achieved")
ax1.legend(*scatter.legend_elements(),title = "Target")
ax2.legend(*scatter.legend_elements(), title = "Target")
ax1.axhline(over50["chol"].mean(),linestyle="--")
ax2.axhline(over50["thalach"].mean(),linestyle="--")
#adding a title
fig.suptitle("Heart Disease Analysis", fontsize = 16 , fontweight = "bold" );
```
